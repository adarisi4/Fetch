```python
def roarin_edge_case_fix(value):
    if 'Roarin' in value:
        value = value.replace('"CAPRI SUN Roarin" Waters Tropical Fruit Juice Drinks 10 ct"', "\"CAPRI SUN Roarin' Waters Tropical Fruit Juice Drinks 10 ct\"")
        value = str(value)
        return value
    else:
        return value
```

```python
receipts['rewardsReceiptItemList'] = receipts['rewardsReceiptItemList'].apply(lambda x: roarin_edge_case_fix(x))
```
